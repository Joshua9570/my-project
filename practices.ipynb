```python
from tkinter import *
from tkinter import ttk
master=Tk()
name_F=Frame(master).grid(row=0,column=0)
Label(name_F,text="Optima Laundry service company").grid(row=0,column=0,columnspan=4)
Label(master,text=" Registration").grid(row=1,column=0)
def submit():
    import sqlite3 as sql
    conn=sql.connect('customer_record.db')
    cur=conn.cursor()
   

    cur.execute("""CREATE TABLE IF NOT EXISTS Customer_database(
                 cus_id INTEGER,
                 surname TEXT,
                 first_name TEXT,
                 last_name TEXT,
                 phone_no TEXT,
                 gender TEXT,
                 address TEXT,
                 status TEXT
                 )
                 """)
    cur.execute("INSERT INTO Customer_database VALUES(:cus_id,:sur_entry, :first_entry, :last_entry,:phone_entry,:gender_entry,:address_entry)",
              {   'cus_id':id_entry.get(),
                  'sur_entry':sur_entry.get(),
                  'first_entry':first_entry.get(),
                  'last_entry':last_entry.get(),
                  'phone_entry':phone_entry.get(),
                  'gender_entry':gender_entry.get(),
                  'address_entry':address_entry.get(),
                  #'status_entry':status_entry.get()
              })
    
    sur_entry.delete(0,End)
    first_entry.delete(0,End)
    last_entry.delete(0,End)
    phone_entry.delete(0,End)
    gender_entry.delete(0,End)
    address_entry.delete(0,End)
    id_entry.delete(0,End)
    conn.commit()
    conn.close()
    
def retrieve():
    import sqlite3 as sql
    conn=sql.connect('customer_record.db')
    cur=conn.cursor()
    
    cur.execute("SELECT * FROM Customer_database")
    result=cur.fetchall()
    for i in result:
        print(i)
        
    
    
    conn.commit()
    conn.close()
    
def new_window():
    reg_window=Tk()
    global sur_entry
    global first_entry
    global last_entry
    global phone_entry
    global status_entry
    global gender_entry
    global address_entry
    global id_entry
    
    cus_id=Label(reg_window,text='Customer ID').grid(row=0,column=0)
    id_entry=Entry(reg_window)
    id_entry.grid(row=0,column=1)
    
    sur_name=Label(reg_window,text='Surname name').grid(row=1,column=0)
    sur_entry=Entry(reg_window)
    sur_entry.grid(row=1,column=1)
    
    first_name=Label(reg_window,text='First name').grid(row=2,column=0)
    first_entry=Entry(reg_window)
    first_entry.grid(row=2,column=1)
    
    last_name=Label(reg_window,text='Last name').grid(row=3,column=0)
    last_entry=Entry(reg_window)
    last_entry.grid(row=3,column=1)
    
    phone_no=Label(reg_window,text='Phone number').grid(row=4,column=0)
    phone_entry=Entry(reg_window)
    phone_entry.grid(row=4,column=1)
    
    gender_name=Label(reg_window,text='Gender').grid(row=5,column=0)
    gender_entry=Entry(reg_window)
    gender_entry.grid(row=5,column=1)
    
    status=Label(reg_window,text='Marital status').grid(row=6,column=0)
    status_entry=Entry(reg_window)
    status_entry.grid(row=6,column=1)
    
    address=Label(reg_window,text='Address').grid(row=7,column=0)
    address_entry=Entry(reg_window)
    address_entry.grid(row=7,column=1)
    
    summit_btn=ttk.Button(reg_window,text='Register',command=submit)
    summit_btn.grid(row=8,column=1)
    
    
reg_btn=ttk.Button(master,text="Click here for Registration",command=new_window)
reg_btn.grid(row=1,column=1,pady=2,padx=2,ipadx=100,columnspan=20)

retrieve_btn=ttk.Button(master,text='List of customers',command=retrieve)
retrieve_btn.grid(row=2,column=0,columnspan=20,ipadx=100)

master.mainloop()
```
